```python
from skimage import measure
import matplotlib as mpl
import cv2
from PIL import Image  # this contain the image class and methods from PIL library
import numpy as np # this imports the numerical and array library 

import matplotlib.pyplot as plt
```

```python
data_dir = r"E:\Python Shit\CS413\coursework-data\lego-bricks\\"
img1 = Image.open(data_dir+"lego-bricks-1.jpg")
img1 = np.asarray(img1)
plt.imshow(img1[:,:,1], cmap="gray")
plt.show()
print(img1[:,:,1].shape)
img1r = img1[:,:,0]
img1g = img1[:,:,1]
img1b = img1[:,:,2]
```

```python
plt.hist(img1r.ravel(), bins=30, range=[0,256]) # ravel() is like flatten()
a = np.asanyarray(img1r.ravel())
mr = a.mean(None)
plt.show()
plt.hist(img1g.ravel(), bins=30, range=[0,256]) # ravel() is like flatten()
a = np.asanyarray(img1g.ravel())
mg = a.mean(None)
plt.show()
plt.hist(img1b.ravel(), bins=30, range=[0,256]) # ravel() is like flatten()
a = np.asanyarray(img1b.ravel())
mb = a.mean(None)
plt.show()
```

```python
chromored = (img1r<=110)
chromogreen = (img1g<=80)
chromoblue = (img1b<=35)
chromobin = np.zeros(img1r.shape, dtype='int')
chromobin[chromored] = 0
chromobin[~chromored] = 1
plt.imshow(chromobin, cmap='gray')
plt.show()

chromobin = np.zeros(img1r.shape, dtype='int')
chromobin[chromoblue] = 0
chromobin[~chromoblue] = 1
plt.imshow(chromobin, cmap='gray')
plt.show()

chromobin = np.zeros(img1r.shape, dtype='int')
chromobin[chromogreen] = 0
chromobin[~chromogreen] = 1
plt.imshow(chromobin, cmap='gray')
plt.show()

chromoall = np.logical_or(np.logical_or(chromored,chromoblue),chromogreen)
chromobin = np.zeros(img1r.shape, dtype='int')
chromobin[chromoall] = 0
chromobin[~chromoall] = 1
plt.imshow(chromobin, cmap='gray')
plt.show()

chroms = np.count_nonzero(chromobin)
tots = img1r.shape[0]*img1r.shape[1]
print((tots-chroms)/tots)
```
